```python
# Importar pandas e numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criar um array aleat√≥rio de 100 valores entre 78 e 85
acuracia = np.random.uniform(78, 85, 100)

# Acur√°cia de cada um dos 100 dados
numeros_de_dados = list(range(1, len(acuracia) + 1))

# Crie o gr√°fico de barras
plt.bar(numeros_de_dados, acuracia)
plt.xlabel('N√∫mero da rodada')
plt.ylabel('Acur√°cia (%)')
plt.title('Acur√°cia de cada dado')
```
