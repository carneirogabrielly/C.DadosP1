```python
# Importar pandas e numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criar um array aleatório de 100 valores entre 78 e 85
acuracia = np.random.uniform(78, 85, 100)

# Acurácia de cada um dos 100 dados
numeros_de_dados = list(range(1, len(acuracia) + 1))

# Crie o gráfico de barras
plt.bar(numeros_de_dados, acuracia)
plt.xlabel('Número da rodada')
plt.ylabel('Acurácia (%)')
plt.title('Acurácia de cada dado')
```

```python
# Importar matplotlib.pyplot
import matplotlib.pyplot as plt

media_aleatoria = 82.2639
media_teste = 83.2
# Criar um array aleatório de 100 valores entre 78 e 85
acuracia = np.random.uniform(80, 85, 100)

# Acurácia de cada um dos 100 dados
numeros_de_dados = list(range(1, len(acuracia) + 1))

# Crie o gráfico de barras
plt.bar(numeros_de_dados, acuracia)
plt.xlabel('Número da rodada')
plt.ylim(77,86)
plt.ylabel('Acurácia (%)')
plt.title('Acurácia de cada dado')

# Adicionar uma linha reta em acurácia = 0.82
plt.axhline(y=media_aleatoria, color='r', linestyle='--')
plt.axhline(y=media_teste, color='k', linestyle='--')

# Mostrar o gráfico
plt.show()
```
